```python
import numpy as np

# Creating a random vector of size 15 with integers of range 1-20 and assigned to the variable called vector
vec = np.random.randint(low=1, high=20, size=15)

# Reshaping the vector to 3 by 5 array
array_3x5 = vec.reshape(3, 5)
print(array_3x5)

# Printing the shape of the array here 
print(array_3x5.shape)

# Replacing the max in each row by 0
array_3x5[np.arange(3), array_3x5.argmax(axis=1)] = 0

# Printing the modified array here
print(array_3x5)

```

```python
import numpy as np

# Defining the input matrix  with some values in it.
Arr1 = np.array([[3, -2], [1, 0]])

# Computing the eigenvalues and right eigenvectors and assigned the variable to it.
eigenvalues, eigenvectors = np.linalg.eig(Arr1)

# Printing the results
print("Eigenvaluesof the input: ", eigenvalues)
print("Right eigenvectors:\n", eigenvectors)

```

```python
import numpy as np

# Defining the input array and assigned values as well
Arr1 = np.array([[0, 1,2 ], [3, 4, 5]])

#  using the tarce() functionality to fingd the sum of the diagonal elements.
sum_values = np.trace(Arr1)

# Printing the result here 
print("Sum of diagonal elements from the array is:",sum_values )

```

```python
import matplotlib.pyplot as plt

# Defining the programming languages and their popularity in the list.
languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

# explode 1st slice
explode = (0.1, 0, 0, 0,0,0)  

# Creating apie chart for this list uisng plt.pie function anad assigned some values in it.
plt.pie(popularity, explode=explode, labels=languages, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)
# Adding titile to the pie chart uisng title function
plt.title("Polpularity Of programming Languages")
plt.axis('equal')
plt.show()

```
